## Step 3: Data Preparation

### Dummy Variables

The variables `season` and `weathersit` has to be converted to dummy variables. 


```python
season = pd.get_dummies(df_bike['season'], drop_first = True, dtype=int)
```

```python
season.rename(columns={2: 'summer', 3: 'fall', 4: 'winter'}, inplace=True)
```

```python
season.head()
```

```python
weather = pd.get_dummies(df_bike['weathersit'], drop_first = True, dtype=int)
```

```python
weather.head()
```

```python
weather.rename(columns={2: 'Misty', 3: 'LightRainSnow'}, inplace=True)
```

```python
weather.head()
```

```python
df_bike = pd.concat([df_bike, season, weather], axis = 1)
```

```python
df_bike.head()
```

```python
# Drop 'season' and 'weathersit' as we have created the dummies for it

df_bike.drop(['season', 'weathersit'], axis = 1, inplace = True)
```

```python
df_bike.head()
```
